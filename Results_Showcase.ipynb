## Key Findings1. **ESM-1b leads** on linear probe accuracy (82.4%), suggesting its 1280-dim embeddings are most linearly separable for functional classification2. **ESM-3 and SaProt** tie on Spearman correlation (0.358), indicating comparable embedding geometry3. **DNA-based models** (AlphaGenome, Evo2) capture distinct information — their representations have low correlation with protein-level models4. **Low sequence identity** between groups validates that models must learn beyond simple sequence matching5. **Statistical significance** confirmed via Friedman test across all models and Wilcoxon signed-rank between top pairs
