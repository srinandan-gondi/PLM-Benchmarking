# Protein Foundation Model Benchmark — Results

**Dataset:** 85 high-quality proteins (pLDDT >= 0.80) across 6 functional groups  
**Models:** ESM-1b, ESM-3, SaProt, AlphaGenome, Evo2 7B

This notebook displays the benchmark results. The full pipeline (requiring GPU) is in `Copy_of_Protein_LLM_Benchmarking_New.ipynb`.

```python
import pandas as pd
from IPython.display import Image, display, HTML
from pathlib import Path

FIGURES = Path("figures")
RESULTS = Path("results")
```

## Classification Benchmark Results

| Metric | Description |
|--------|-------------|
| Spearman | Correlation between embedding cosine similarity and same-group membership |
| LinearProbe | 3-fold CV accuracy with logistic regression |
| kNN | 3-fold CV accuracy with k=5 nearest neighbors (cosine) |
| StructuralSpearman | Correlation with experimental 3D structure similarity |

```python
df = pd.read_csv(RESULTS / "final_benchmark_metrics.csv")
df.style.format({
    "Spearman": "{:.3f}",
    "LinearProbe": "{:.2%}",
    "kNN": "{:.2%}",
    "StructuralSpearman": "{:.3f}"
}).background_gradient(cmap="Greens", subset=["LinearProbe", "kNN"])
```

## Key Findings

1. **ESM-1b leads** on linear probe accuracy (82.4%), suggesting its 1280-dim embeddings are most linearly separable for functional classification
2. **ESM-3 and SaProt** tie on Spearman correlation (0.358), indicating comparable embedding geometry
3. **DNA-based models** (AlphaGenome, Evo2) capture distinct information — their representations have low correlation with protein-level models
4. **Low sequence identity** between groups validates that models must learn beyond simple sequence matching
5. **Statistical significance** confirmed via Friedman test across all models and Wilcoxon signed-rank between top pairs
